## Kolommen verwijderen

```python
product.drop('LANGUAGE', axis = 1)
```

We ontdekten hierv√≥√≥r al dat de kolom 'LANGUAGE' eigenlijk overbodig is, aangezien hier continu 'EN' in staat. Voor analyse is dit dus niet heel relevant. We besluiten daarom om deze kolom te verwijderen met bovenstaande code. 'axis = 1' specificeert dat hij de gegeven te verwijderen string moet zoeken in de kolomnamen, en niet in de rijen van het betreffende dataframe.

```python
product
```

Huh? Dit is gek! We hadden 'LANGUAGE' toch verwijderd? Hij staat er nu gewoon weer! Hoe kan dat?
Antwoord: in het codeblok waarin we '.drop(...)' voor het eerst toepasten hebben we slechts een tijdelijke view gecre√´erd op het dataframe. Dat wil zeggen dat als je dit frame opnieuw aanroept je de oorspronkelijke versie ervan krijgt (dus inclusief 'LANGUAGE', voordat we deze kolom weghaalden). Dat willen we niet. Oplossing: de gecre√´erde view toewijzen aan het oorspronkelijke DataFrame.

```python
product = product.drop('LANGUAGE', axis = 1)
product
```

Nice! Met de "DataFrame = DataFrame"-notatie gaat het wel goed. Vanaf nu krijgen we het ge√ºpdate DataFrame als we "product" aanriepen.

## ORDER BY

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product ORDER BY production_cost DESC". Let erop dat als je aflopend wilt sorteren, je 'ascending = False' schrijft in de tweede parameter van sort_values(). Als je oplopend wilt sorteren hoef je geen tweede parameter in te vullen.

```python
product.sort_values(['PRODUCTION_COST', 'MARGIN'], ascending = [False, True])
```

Als je de ene kolom aflopend wilt laten sorteren en de andere oplopend: let dan goed op de booleans tussen blokhaken in de tweede parameter van sort_values(). Deze moeten matchen met de kolomnamen tussen blokhaken in de eerste parameter van sort_values (reminder: True is oplopend, False is aflopend).

## WHERE

### Exacte voorwaarden zonder NULL

```python
production_cost_more_than_350 = product['PRODUCTION_COST'] > 350
product.loc[production_cost_more_than_350, :]
```

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product WHERE production_cost > 350". Eindelijk kunnen we de eerste parameter van .loc[] invullen. Dit doen we met een boolean...
- We herhalen de naam van het gecre√´erde dataframe ('product').
- Tussen blokhaken specificeren we de naam van de kolom waarop we willen filteren.
- We geven met een vergelijkingsoperator (>) en een getal (350) aan welke getallen we wel en niet willen zien.

In de tweede parameter van .loc[] staat :. Reminder: hiermee specificeren we dat we alle kolommen willen zien. We filteren in dit geval dus alleen op rijen (eerste parameter) en niet op kolommen (tweede parameter).

```python
product.loc[product['PRODUCT_NUMBER'].isna(), :]
#Geeft alle kolommen (':') van alle producten waarvan het PRODUCT_NUMBER onbekend is (kennelijk is dat nooit het geval)
```

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product WHERE PRODUCT_NUMBER IS NULL". Om op NULL-waarden te checken cre√´ren we een boolean door na de blokhaken een functie aan te roepen (.isna()). Kennelijk is het PRODUCT_NUMBER altijd bekend. De tegenhanger van .isna() is trouwens .notna() (vergelijkbaar met "NOT NULL" in SQL).

### Niet-exacte voorwaarden met LIKE

```python
product.dtypes
```

```python
product.loc[product['PRODUCT_NUMBER'].str[0] == '1', :]
```

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product WHERE PRODUCT_NAME LIKE 'B%'". Ditmaal cre√´ren we de boolean in de eerste parameter van .loc[] met een speciaal attribuut (.str[]). In de blokhaken van str specificeren we met 0 dat alleen de eerste letter van elke waarde van PRODUCT_NAME gecontroleerd moet worden. Index = 0 is namelijk gelijk aan de eerste letter, met index = 1 pak je de tweede letter, enzovoort.

```python
product.loc[product['PRODUCT_NAME'].str[0:4] == 'Fire', :]
```

```python
product.loc[product['PRODUCT_NAME'].str[-1] == 'n', :]
```

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product WHERE PRODUCT_NAME LIKE '%n'". We voeren nu in .str de index '-1' in. Dit is op het eerste gezicht misschien gek, maar Python staat 'reverse indexing' toe, ten teken dat je als index in reeksen (en waarden z√≠jn reeksen) ook mingetallen mag uitvoeren. Met index = -1 pak je de laatste letter, met index = -2 pak je de een-na-laatste letter, enzovoort.

```python
product.loc[product['PRODUCT_NAME'].str.contains('e'), :]
```

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product WHERE PRODUCT_NAME LIKE '%e%'". In plaats van de blokhaken na str in te vullen roepen we nu een functie aan (.contains). Hierin specificeren we een mogelijke invulling die overal in de betreffende rijwaarden voor mag komen (begin, eind, midden; het maakt niet uit).

### Voorwaarden met logische operatoren (AND & OR)

```python
product_name_begins_with_b = product['PRODUCT_NAME'].str[0] == 'B'
product_name_ends_with_s = product['PRODUCT_NAME'].str[-1] == 's'

product.loc[(product_name_begins_with_b) & (product_name_ends_with_s), :]
```

Dit is het Pythonequivalent van de SQL-query: "SELECT * FROM product WHERE PRODUCT_NAME LIKE 'B%' OR PRODUCT_NAME LIKE '%s%'". In de eerste parameter van .loc[] cre√´ren we 2 booleans, elk van hen wordt gescheiden door ronde haakjes en een logische operator. 2 keuzes:
- & voor 'AND' (alle booleans moeten waar zijn, wil de rij in de selectie opgenomen worden).
- | voor 'OR' (slechts 1 boolean hoeft waar te zijn, wil de rij in de selectie opgenomen worden).

## GROUP BY

```python
grouped = product.groupby('PRODUCT_TYPE_CODE')['PRODUCTION_COST'].mean().reset_index()
grouped = grouped.rename(columns = {'PRODUCTION_COST' : 'AVG_PRODUCTION_COST'})

grouped
```

Dit is het Pythonequivalent van de SQL-query: "SELECT PRODUCT_TYPE_CODE, AVG(PRODUCTION_COST) FROM product GROUP BY PRODUCT_TYPE_CODE". De rename is nodig, omdat anders de kolomnaam "PRODUCTION_COST" letterlijk blijft staan. Voor de raadpleger van dit dataframe is het wel zo aardig om duidelijk te maken dat het om een gemiddelde gaat, dus maken we hier "AVG_PRODUCTION_COST" van. Merk op dat je hier eerst een dataframe 'grouped' aanmaakt en deze vervolgens overschrijft door de rename hieraan toe te voegen: met 'grouped = grouped' wijzig je dus het oorspronkelijke dataframe.

```python
grouped.loc[grouped['AVG_PRODUCTION_COST'] < 100, :]
```

Dit is het Pythonequivalent van de SQL-query: "SELECT PRODUCT_TYPE_CODE, AVG(PRODUCTION_COST) FROM product GROUP BY PRODUCT_TYPE_CODE HAVING AVG(PRODUCTION_COST) < 100". We hadden de GROUP BY al toegepast, het dataframe 'grouped' is daar het resultaat van. De HAVING realiseren we door .loc[] op grouped toe te passen en de eerste parameter daarvan in te vullen met een boolean.

## JOIN

```python
order_details = pd.read_sql("SELECT * FROM order_details", sales_conn)
order_details
```

Om de JOIN te demonstreren (zie hieronder) wordt van een tweede databasetabel (order_details) een dataframe gemaakt.

```python
joined = pd.merge(product, order_details, on = 'PRODUCT_NUMBER', how = 'inner')
joined
```

Dit is het Pythonequivalent van de SQL-query: "SELECT od.order_detail_code, p.product_type_code FROM product p JOIN order_details od ON p.product_number = od.product_number". Als je meerdere joins wilt uitvoeren zul je je reeds gemaakt 'joined'-dataframe moeten overschrijven. Met how = 'inner' geef je ten slotte aan dat het om een INNER JOIN gaat, je kan hier ook 'left' (voor een LEFT JOIN), 'right' (voor een RIGHT JOIN) en 'outer' (voor een OUTER JOIN) schrijven.

## IIF & CASE

```python
for index, row in product.iterrows():
    production_cost = product.at[index, 'PRODUCTION_COST']
    
    if(production_cost < 100):
        product.at[index, 'MARGIN_RANKING'] = 'cheap'
    elif(production_cost < 200):
        product.at[index, 'MARGIN_RANKING'] = 'a bit expensive'
    else:
        product.at[index, 'MARGIN_RANKING'] = 'very expensive'

product
```

De code-loop hierboven loopt elke rij langs van het DataFrame "Product". Bij elke rij die langsgelopen wordt krijgt de variabele "index" de indexwaarde (zie de dikgedrukte getallen in de eerste kolom). Met behulp van .at[] kan je op deze index en een gespecificeerde kolom een bepaalde waarde ophalen en deze ook (conditioneel) wijzigen. Het mooie hiervan? Als de gespecificeerde kolom niet bestaat (wat in dit geval zo is bij 'MARGIN_RANKING') maakt Python die kolom zelf aan, als laatste kolom in het DataFrame.
